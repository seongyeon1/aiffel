## ì„  ê·¸ë˜í”„ ê·¸ë ¤ë³´ê¸°
ì´ë²ˆì—ëŠ” matplotlibì„ ì´ìš©í•´ ì„ (line) ê·¸ë˜í”„ë¥¼ ê·¸ë ¤ë³´ê² ìŠµë‹ˆë‹¤.

### ë°ì´í„° ì •ì˜
ìš°ì„  ëª¨ë“ˆì„ import í•˜ê³  ê·¸ë˜í”„ë¡œ ê·¸ë¦´ ë°ì´í„°ë¥¼ ì •ì˜í•´ ì¤ë‹ˆë‹¤. ì‚¬ìš©í•  ë°ì´í„°ëŠ” ê³¼ê±° ì•„ë§ˆì¡´ ì£¼ê°€ ë°ì´í„°ì…ë‹ˆë‹¤.

- https://finance.yahoo.com/quote/AMZN/history?p=AMZN     
í„°ë¯¸ë„ì—ì„œ í´ë¼ìš°ë“œì— ë¯¸ë¦¬ ì €ì¥ëœ íŒŒì¼ì— ì‹¬ë³¼ë¦­ ë§í¬ë¥¼ ê±¸ì–´ ë°ì´í„°ë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆë„ë¡ í•©ë‹ˆë‹¤.

ì ì´ë²ˆì—” ì£¼ì„(annotation) ë“± ì¢€ ë” ê³ ê¸‰ ê¸°ë²•ì„ í™œìš©í•˜ì—¬ ê·¸ë˜í”„ë¥¼ ê·¸ë ¤ ë³¼ê¹Œìš”?

````python
from datetime import datetime
import pandas as pd
import os

# ê·¸ë˜í”„ ë°ì´í„° 
csv_path = os.getenv("HOME") + "/aiffel/data_visualization/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']

# ì¶• ê·¸ë¦¬ê¸° ë° ì¢Œí‘œì¶• ì„¤ì •
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600, 2200])
plt.xlim(['2019-05-01', '2020-03-01'])

# ì£¼ì„ë‹¬ê¸°
important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020, 2, 19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d)+10), # ì£¼ì„ì„ ë‹¬ ì¢Œí‘œ(x,y)
                xytext=(d, price.asof(d)+100), # ì£¼ì„ í…ìŠ¤íŠ¸ê°€ ìœ„ì¹˜í•  ì¢Œí‘œ(x,y)
                arrowprops=dict(facecolor='red')) # í™”ì‚´í‘œ ì¶”ê°€ ë° ìƒ‰ ì„¤ì •

# ê·¸ë¦¬ë“œ, íƒ€ì´í‹€ ë‹¬ê¸°
plt.grid()
ax.set_title('StockPrice')

# ë³´ì—¬ì£¼ê¸°
plt.show()
````

### ê·¸ë¦¬ë“œ
`grid()` ë©”ì„œë“œë¥¼ ì´ìš©í•˜ë©´ ê·¸ë¦¬ë“œ(ê²©ìëˆˆê¸ˆ)ë¥¼ ì¶”ê°€í•  ìˆ˜ ìˆì–´ìš”.

````python
# Q. ë‚ ì§œë³„ ì¢…ê°€(Close)ê°€ ì•„ë‹Œ ìµœê³ ê°€(High) ë°ì´í„°ë¥¼ ì´ìš©í•´ì„œ ìœ„ì™€ ê°™ì€ ê·¸ë˜í”„ë¥¼ ê·¸ë ¤ë³´ì„¸ìš”!

# ê·¸ë˜í”„ ë°ì´í„° 
csv_path = os.getenv("HOME") + "/aiffel/data_visualization/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['High']

# ì¶• ê·¸ë¦¬ê¸° ë° ì¢Œí‘œì¶• ì„¤ì •
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600, 2200])
plt.xlim(['2019-05-01','2020-03-01'])

# ë‚ ì§œ ì •ë³´ë¥¼ python datetime ìë£Œí˜•ìœ¼ë¡œ ë³€í™˜
min_date = price.idxmin().to_pydatetime()
max_date = price.idxmax().to_pydatetime()

# ì£¼ì„ë‹¬ê¸°
important_data = [(min_date, "Low Price"), (max_date, "Peak Price")]
for d, label in important_data: # d : ìµœì†Œê°’ì˜ ë‚ ì§œ, labelì€ ê¸€ì
    ax.annotate(label, xy=(d, price.asof(d)+10), # ì£¼ì„ì„ ë‹¬ ì¢Œí‘œ(x,y)
                xytext=(d, price.asof(d)+100),  # ì£¼ì„ í…ìŠ¤íŠ¸ê°€ ìœ„ì¹˜í•  ì¢Œí‘œ(x,y)
                arrowprops=dict(facecolor='red')) # í™”ì‚´í‘œ ì¶”ê°€ ë° ìƒ‰ ì„¤ì •

# ê·¸ë¦¬ë“œ, íƒ€ì´í‹€ ë‹¬ê¸°
plt.grid()
ax.set_title('StockPrice')

# ë³´ì—¬ì£¼ê¸°
plt.show()
````

### linestyle, marker ì˜µì…˜
ë¼ì¸ ìŠ¤íƒ€ì¼ì€ plot()ì˜ ì¸ìë¡œ ë“¤ì–´ê°€ëŠ”ë°ìš”, ì•„ë˜ì™€ ê°™ì´ ë‹¤ì–‘í•œ ë°©ë²•ìœ¼ë¡œ í‘œê¸°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë‹¤ìŒê³¼ ê°™ì€ ë°ì´í„°ì˜ íˆìŠ¤í† ê·¸ë¨ì„ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

x1ì€ í‰ê· ì€ 100ì´ê³  í‘œì¤€í¸ì°¨ëŠ” 15ì¸ ì •ê·œë¶„í¬ë¥¼ ë”°ë¦…ë‹ˆë‹¤.    
x2ëŠ” í‰ê· ì€ 130ì´ê³  í‘œì¤€í¸ì°¨ëŠ” 15ì¸ ì •ê·œë¶„í¬ë¥¼ ë”°ë¦…ë‹ˆë‹¤.    
ë„ìˆ˜ë¥¼ 50ê°œì˜ êµ¬ê°„ìœ¼ë¡œ í‘œì‹œí•˜ë©°, í™•ë¥  ë°€ë„ê°€ ì•„ë‹Œ ë¹ˆë„ë¡œ í‘œê¸°í•©ë‹ˆë‹¤.

````python
#ê·¸ë˜í”„ ë°ì´í„°
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

# ì¶• ê·¸ë¦¬ê¸°
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
patches = ax1.hist(x1, bins=50, density=False)  # binsëŠ” xê°’ì„ ì´ 50ê°œ êµ¬ê°„ìœ¼ë¡œ ë‚˜ëˆˆë‹¤ëŠ” ëœ»ì…ë‹ˆë‹¤.
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)
ax1.xaxis.set_ticks_position('bottom')  # xì¶•ì˜ ëˆˆê¸ˆì„ ì•„ë˜ í‘œì‹œ 
ax1.yaxis.set_ticks_position('left')  # yì¶•ì˜ ëˆˆê¸ˆì„ ì™¼ìª½ì— í‘œì‹œ

# ë¼ë²¨, íƒ€ì´í‹€ ë‹¬ê¸°
plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
ax1.set_title('Two Frequency Distributions')

# ë³´ì—¬ì£¼ê¸°
plt.show()
````
